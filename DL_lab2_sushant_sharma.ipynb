```python
X=[0.75,0.65,0.80]
w=[0.4,0.3,0.5]
b=-0.2
Y=1
n=0.01
```

```python
Z=(w[0]*X[0])+(w[1]*X[1])+(w[2]*X[2])+b
print(Z)
```

```python
def step_function(z):
  if z>=0: return 1
  return 0
```

```python
y_pred=step_function(Z)
print(y_pred)
```

```python
error=Z-y_pred
error
```

```python
if error != 0:
    for i in range(len(w)):
        w[i] = w[i] + (n * error * X[i])

    print("\nStep 5: Updated Weights")
    print("New Weights =", w)
else:
    print("\nNo Weight Update Needed")
```

```python
b = b + (n * error)

print("\nStep 6: Updated Bias")
print("New Bias =", b)
```

```python
z_new = 0
for i in range(len(X)):
    z_new = z_new + (X[i] * w[i])

z_new = z_new + b

if z_new >= 0:
    y_new = 1
else:
    y_new = 0
```

```python

print("\nStep 7: After Training")
print("New z =", z_new)
print("New Prediction =", y_new)


# Final Result

print("\nFinal Result:")
if y_new == Z:
    print("Target Achieved (Pass = 1)")
else:
    print("Target Not Achieved")
```
